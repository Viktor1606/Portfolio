```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=15,n_features=5, class_sep=0.2, weights=[0.6],random_state=42)
```

```python
df = pd.DataFrame(X)
```

```python
df['y_type'] = y
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(df[[0,1,2,3,4]], df['y_type'])
```

```python
lr.predict(df[[0,1,2,3,4]])
```

```python
lr.predict_proba(df[[0,1,2,3,4]])[:,1]
```

```python
df['y_score'] = lr.predict_proba(df[[0,1,2,3,4]])[:,1]
```

```python
df
```

```python
df['y_pred'] = (df['y_score'] > 0.5)*1
```

```python
df
```

```python
def get_cm(row):
  if row['y_type'] == 1 and row['y_pred'] ==1:
    return 'tp'
  if row['y_type'] == 0 and row['y_pred'] ==0:
    return 'tn'
  if row['y_type'] == 1 and row['y_pred'] ==0:
    return 'fn'
  if row['y_type'] == 0 and row['y_pred'] ==1:
    return 'fp'
```

```python
df['cm']= df.apply(get_cm, axis=1)
```

```python
tp,tn,fp,fn
```

```python
res = []
for thr in thrs:
  df['y_pred'] = (df['y_score'] > thr)*1
  tp = len(df[(df['y_type']== 1) & (df['y_pred'] == 1)])
  tn = len(df[(df['y_type']== 0) & (df['y_pred'] == 0)])
  fp = len(df[(df['y_type']== 0) & (df['y_pred'] == 1)])
  fn = len(df[(df['y_type']== 1) & (df['y_pred'] == 0)])
  res.append((tp,tn,fp,fn,thr))
```

```python
cm = pd.DataFrame(res)
```

```python
'tp,tn,fp,fn,thr'.split(',')
```

```python
cm.columns = ['tp', 'tn', 'fp', 'fn', 'thr']
```

```python
cm
```

```python
cm['tpr'] = cm['tp'] / (cm['tp'] + cm['fn'])
```

```python
cm['fpr'] = cm['fp'] / (cm['fp'] + cm['tn'])
```
